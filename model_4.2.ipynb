```python
import numpy as np
import pandas as pd
import os

from sklearn.model_selection import train_test_split

from sklearn.model_selection import KFold, cross_val_score
from sklearn.pipeline import make_pipeline
from sklearn.kernel_ridge import KernelRidge
from sklearn.preprocessing import RobustScaler
```

```python
# 读取数据
import pandas as pd
da34 = pd.read_csv("zhuhaihouse_chuli_34.csv")
print(da34)
```

```python
da34= da34.drop(["街区","街区目标编码","租赁方式","朝向","朝向编码","行政区","板块","房屋面积","格局","电梯","车位","用水","用电","燃气","采暖"], axis=1)
```

```python
da34 = pd.DataFrame(da34)
da34
```

```python
labels = da34['每月租金']
features = da34.drop("每月租金", axis=1)
X_train,X_test,y_train,y_test= train_test_split(features,labels,test_size=0.3,random_state=0)
X_train
```

```python
y_test = pd.DataFrame(data=y_test, columns=["每月租金"])
y_test_old = y_test
y_test_old
```

```python
y_train = pd.DataFrame(data=y_train, columns=["每月租金"])
y_train
```

```python
y_max = y_train.max(axis=0)
y_max
```

```python
y_min = y_train.min(axis=0)
y_min
```

```python
from sklearn import preprocessing
x_scaler = preprocessing.MinMaxScaler(feature_range=(0, 1))
y_scaler = preprocessing.MinMaxScaler(feature_range=(0, 1))
#先归一化训练数据
X_train = x_scaler.fit_transform(X_train)
y_train = y_scaler.fit_transform(y_train)
#再来归一化测试数据
X_test = x_scaler.transform(X_test)
y_test = y_scaler.transform(y_test)
```

```python
X_train = pd.DataFrame(data=X_train, columns=["楼层","街区目标编码_m5","朝向编码二元化", "房屋面积数值","室","厅","卫","电梯编码","车位编码","用水编码","用电编码","燃气编码"])
X_train
```

```python
X_test = pd.DataFrame(data=X_test, columns=["楼层","街区目标编码_m5","朝向编码二元化", "房屋面积数值","室","厅","卫","电梯编码","车位编码","用水编码","用电编码","燃气编码"])
X_test
```

```python
y_train = pd.DataFrame(data=y_train, columns=["每月租金"])
y_train
```

```python
y_test = pd.DataFrame(data=y_test, columns=["每月租金"])
y_test
```

```python
# 定义一个交叉评估函数 Validation function
n_folds = 5
def rmsle_cv(model):
    kf = KFold(n_folds, shuffle=True, random_state=42).get_n_splits(X_train.values)
    rmse= np.sqrt(-cross_val_score(model, X_train.values,y_train , scoring="neg_mean_squared_error", cv = kf))
    return(rmse)

# 岭回归（Kernel Ridge Regression）  
KRR = make_pipeline(RobustScaler(), KernelRidge(alpha=0.6, kernel='polynomial', degree=2, coef0=2.5))
score = rmsle_cv(KRR)
print("\nKRR score: {:.4f} ({:.4f})\n".format(score.mean(), score.std())) 


''' 
预测
'''
import time 
KRR.fit(X_train.values,y_train.values)
tic=time.perf_counter()
y = KRR.predict(X_test.values)
toc=time.perf_counter()
print("elapsed time:",toc-tic)
```

```python
from sklearn.metrics import mean_squared_error as MSE 
print("MSE:",MSE(y_test,y))

from sklearn.metrics import mean_absolute_error as MAE 
print("MAE:",MAE(y_test,y))

from sklearn.metrics import r2_score
print("R2:",r2_score(y_test,y))

from sklearn.metrics import explained_variance_score as EVC 
print("EVC:",EVC(y_test,y))
```

```python
from sklearn.linear_model import Lasso
from sklearn.metrics import r2_score
lasso = Lasso()
y_pred_lasso = lasso.fit(X_train, y_train)
tic=time.perf_counter()
y_pred_lasso=y_pred_lasso.predict(X_test)
toc=time.perf_counter()
print("elapsed time:",toc-tic)
```

```python
from sklearn.metrics import mean_squared_error as MSE 
print("MSE:",MSE(y_test,y_pred_lasso))

from sklearn.metrics import mean_absolute_error as MAE 
print("MAE:",MAE(y_test,y_pred_lasso))

from sklearn.metrics import r2_score
print("R2:",r2_score(y_test,y_pred_lasso))

from sklearn.metrics import explained_variance_score as EVC 
print("EVC:",EVC(y_test,y_pred_lasso))
```
