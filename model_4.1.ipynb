```python
X_test = pd.DataFrame(data=X_test, columns=["楼层","街区目标编码_m5","朝向编码二元化", "房屋面积数值","室","厅","卫","电梯编码","车位编码","用水编码","用电编码","燃气编码"])
X_test
```

```python
y_train = pd.DataFrame(data=y_train, columns=["每月租金"])
y_train
```

```python
y_test = pd.DataFrame(data=y_test, columns=["每月租金"])
y_test
```

```python
# 定义一个交叉评估函数 Validation function
n_folds = 5
def rmsle_cv(model):
    kf = KFold(n_folds, shuffle=True, random_state=42).get_n_splits(X_train.values)
    rmse= np.sqrt(-cross_val_score(model, X_train.values,y_train , scoring="neg_mean_squared_error", cv = kf))
    return(rmse)

# 岭回归（Kernel Ridge Regression）  
KRR = make_pipeline(RobustScaler(), KernelRidge(alpha=0.6, kernel='polynomial', degree=2, coef0=2.5))
score = rmsle_cv(KRR)
print("\nKRR score: {:.4f} ({:.4f})\n".format(score.mean(), score.std())) 


''' 
预测
'''
import time 
KRR.fit(X_train.values,y_train.values)
tic=time.perf_counter()
y = KRR.predict(X_test.values)
toc=time.perf_counter()
print("elapsed time:",toc-tic)
```

```python
from sklearn.metrics import mean_squared_error as MSE 
print("MSE:",MSE(y_test,y))

from sklearn.metrics import mean_absolute_error as MAE 
print("MAE:",MAE(y_test,y))

from sklearn.metrics import r2_score
print("R2:",r2_score(y_test,y))

from sklearn.metrics import explained_variance_score as EVC 
print("EVC:",EVC(y_test,y))
```

```python
from sklearn.linear_model import Lasso
from sklearn.metrics import r2_score
lasso = Lasso()
y_pred_lasso = lasso.fit(X_train, y_train)
tic=time.perf_counter()
y_pred_lasso=y_pred_lasso.predict(X_test)
toc=time.perf_counter()
print("elapsed time:",toc-tic)
```

```python
from sklearn.metrics import mean_squared_error as MSE 
print("MSE:",MSE(y_test,y_pred_lasso))

from sklearn.metrics import mean_absolute_error as MAE 
print("MAE:",MAE(y_test,y_pred_lasso))

from sklearn.metrics import r2_score
print("R2:",r2_score(y_test,y_pred_lasso))

from sklearn.metrics import explained_variance_score as EVC 
print("EVC:",EVC(y_test,y_pred_lasso))
```
